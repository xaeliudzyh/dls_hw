```python
cat_cols_to_encode = [
    'HasMultiplePhoneNumbers',
    'HasInternetService',
    'HasOnlineSecurityService',
    'HasOnlineBackup',
    'HasDeviceProtection',
    'HasTechSupportAccess',
    'HasOnlineTV',
    'HasMovieSubscription',
    'HasContractPhone',
    'PaymentMethod'
]

encoded_cols = pd.get_dummies(data[cat_cols_to_encode])
data_orig=data
data_encoded = data.drop(cat_cols_to_encode, axis=1)
data = pd.concat([data_encoded, encoded_cols], axis=1)

test_encoded_cols = pd.get_dummies(test_data[cat_cols_to_encode])
test_data_orig = test_data
test_data_encoded = test_data.drop(cat_cols_to_encode, axis=1)
test_data = pd.concat([test_data_encoded, test_encoded_cols], axis=1)

data['Sex']=data['Sex'].apply(lambda x: 1 if x == 'Male' else 0)
test_data['Sex']=test_data['Sex'].apply(lambda x: 1 if x == 'Male' else 0)
data['HasPartner']=data['HasPartner'].apply(lambda x: 1 if x == 'Yes' else 0)
test_data['HasPartner']=test_data['HasPartner'].apply(lambda x: 1 if x == 'Yes' else 0)
data['HasChild']=data['HasChild'].apply(lambda x: 1 if x == 'Yes' else 0)
test_data['HasChild']=test_data['HasChild'].apply(lambda x: 1 if x == 'Yes' else 0)
data['IsBillingPaperless']=data['IsBillingPaperless'].apply(lambda x: 1 if x == 'Yes' else 0)
test_data['IsBillingPaperless']=test_data['IsBillingPaperless'].apply(lambda x: 1 if x == 'Yes' else 0)
data.head(3)
```

```python
preprocessor = ColumnTransformer(
    transformers=[('scaler',
                  Pipeline([
    ('powerTransformer', PowerTransformer(method='yeo-johnson', standardize=False)),
    ('robust', RobustScaler())
    ]), num_cols)],
    remainder='passthrough',
)

log_pipe = Pipeline([
    ('normalizer', preprocessor),
    ('clf', LogisticRegressionCV(
        cv=5,
        penalty='l2',
        solver='liblinear',
        scoring='roc_auc',
        max_iter=1000,
        n_jobs=-1,
        random_state=42,
    ))
])



log_pipe.fit(X_train, y_train)

print("train score:", log_pipe.score(X_train, y_train))
print("test score:", log_pipe.score(X_test, y_test))
print("params:", log_pipe.named_steps['clf'].get_params())
```

```python
from catboost import CatBoostClassifier, Pool
cat_cols = [
    'Sex',
    'IsSeniorCitizen',
    'HasPartner',
    'HasChild',
    'HasPhoneService',
    'HasMultiplePhoneNumbers',
    'HasInternetService',
    'HasOnlineSecurityService',
    'HasOnlineBackup',
    'HasDeviceProtection',
    'HasTechSupportAccess',
    'HasOnlineTV',
    'HasMovieSubscription',
    'HasContractPhone',
    'IsBillingPaperless',
    'PaymentMethod'
]
data_cat = pd.read_csv('train.csv')
X_train, X_test, y_train, y_test = train_test_split(data_cat.drop(target_col,axis=1), data_cat[target_col], train_size=0.8, random_state=42)
param_grid = {
    'iterations':    [300, 500, 1000],
    'depth':         [4, 6],
    'learning_rate': [0.01, 0.03, 0.1],
    'l2_leaf_reg':   [1, 3, 5]
}

train_orig_pool = Pool(X_train, y_train, cat_features=cat_cols)

clf_orig_data = CatBoostClassifier(random_state=42, cat_features=cat_cols,eval_metric='AUC' , verbose=100)

clf_orig_data.grid_search(
    param_grid,
    X=train_orig_pool,
    cv=5,
    stratified=True,
    plot=True,
    search_by_train_test_split = False,
    verbose=100
)
```

```python
y_pred_orig=clf_orig_data.predict(X_test)
clf_orig_data.score(y_pred_orig, y_test)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop(target_col,axis=1), data[target_col], train_size=0.8, random_state=42, stratify=data[target_col])
preprocessor = ColumnTransformer(
    transformers=[('scaler',
                  Pipeline([
    ('powerTransformer', PowerTransformer(method='yeo-johnson', standardize=False)),
    ('robust', RobustScaler())
    ]), num_cols)],
    remainder='passthrough',
)
pipe_cat = Pipeline(
    [('normalizer', preprocessor),
     ('clf', CatBoostClassifier(random_seed=42, task_type='GPU',verbose=2))
])

param_grid = {
    'clf__iterations':    [300, 500, 1000],
    'clf__depth':         [4, 6, 8, 10],
    'clf__learning_rate': [0.01, 0.03, 0.1],
    'clf__l2_leaf_reg':   [1, 3, 5]
}

grid_mod = GridSearchCV(
    estimator=pipe_cat,
    param_grid=param_grid,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=2
)
grid_mod.fit(X_train, y_train)
```

```python
print('params', grid_mod.best_params_)
print('score', grid_mod.best_score_)
print('test score',grid_mod.score(X_test, y_test))
```
